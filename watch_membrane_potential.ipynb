```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from matplotlib.widgets import Button

data = np.load("C:\\Users\\CaoSansheng\\Desktop\\CMSA\\Bump_example.npy")

for i in range(50,70,5):
    v = data[:,i,55]
    plt.plot(range(v.shape[0]),v,label=f'({i},{55})')
plt.xlabel('t/ms')
plt.ylabel('v/mV')
plt.legend()
plt.show()

plt.imshow(data[0,:,:],cmap='coolwarm')
plt.show()
```
